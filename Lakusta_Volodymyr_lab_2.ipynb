# Ініціалізація датасету


```
import pandas as pd

url = "https://raw.githubusercontent.com/vovachuck/EMPI/master/Video_Games_Sales_as_at_22_Dec_2016.csv"
df = pd.read_csv(url)
df['User_Count'] = pd.to_numeric(df['User_Count'], errors='coerce')
df.head()
```

## Bar Plots

```
import matplotlib.pyplot as plt
x1 = [1, 3, 4, 5, 6, 7, 9]
y1 = [4, 7, 2, 4, 7, 8, 3]

x = df.User_Count




plt.bar(range(0,len(x)),x, label="User_Count", color='b')

plt.plot()

plt.xlabel("user length")
plt.ylabel("user height")
plt.title("user Chart ")
plt.legend()
plt.show()
```

##Boxplot

```
df.boxplot(column=['User_Count'])
```

## Density Chart

```
import seaborn as sns

sns.kdeplot(df['User_Count'], shade=True)
```

##Deplot

```
# library & dataset
import seaborn as sns


sns.kdeplot(df['User_Count'], shade=True)
sns.plt.show()
```

##Pie chart

```
import matplotlib.pyplot as plt

labels = '192', '500', '800'
sections = [len(df[df.User_Count == 192].index),
            len(df[df.User_Count == 322].index),
            len(df[df.User_Count == 709].index)]
colors = ['#ed3a01', '#21e12c', '#0095aa']

plt.pie(sections, labels=labels, colors=colors,
        startangle=90,
        autopct = '%1.2f%%')

plt.axis('equal')

plt.show()
```

## Circle chart

```
import matplotlib.pyplot as plt
 


dfList1 = [len(df[df.User_Count <= 192.0]),
            len(df[df.User_Count >= 322.0]),
            len(df[df.User_Count > 709.0])] 

plt.pie(dfList1)

 
# add a circle at the center
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
 
plt.show()
#зелений - значення більше 322
#синій - значення більше 709
#червоний - значення менше 192 
```

## Hist chart

```
import numpy as np
import matplotlib.pyplot as plt

h = df.Rating 
plt.hist(h, range = (1,5))
plt.title("User_Count")
plt.xticks(np.arange(1,5, step=0.5)+0.5, range(1,5))
plt.show()
```

##StackPlot

```
import matplotlib.pyplot as plt


dfList1 = df['User_Count'].tolist()



plt.plot([], [], color='r', label = 'D 1')


plt.stackplot(range(0,len(dfList1)),  dfList1, colors= ['r'])
plt.title('Stack Plot Example')
plt.legend()
plt.show()
```

##Violinplot chart

```
import seaborn as sns

sns.violinplot( y=df["User_Count"] )
sns.plt.show()
```
